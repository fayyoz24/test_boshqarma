```python
# Import necessary libraries
import json
import django
import os
from django.core.wsgi import get_wsgi_application

# Setup Django environment
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'IT_KLASTER/klaster/klaster/settings.py')  # Replace with your project name
django.setup()

# Import your models
from test_olish.models import Question, Option, Theme  # Replace 'your_app' with your app name
from corecode.models import Subject, Class
# Your JSON data (you can also load this from a file)
questions_data = questions_data

def import_questions():
    subject = Subject.objects.get_or_create(name = "Ona tile")
    class_ = Class.objects.get_or_create(name = "5 A")
    # Get or create a theme
    theme, created = Theme.objects.get_or_create(name="Fonetika", 
    subject=subject, class_name=class_)  # Replace with your desired theme name
    
    # Counter for tracking progress
    questions_created = 0
    options_created = 0
    
    try:
        # Iterate through each question in the data
        for q_data in questions_data['questions']:
            # Create question
            question = Question.objects.create(
                theme=theme,
                title=q_data['question']
            )
            questions_created += 1
            
            # Create options for the question
            for opt in q_data['options']:
                for letter, content in opt.items():
                    if letter in ['A', 'B', 'C', 'D']:
                        Option.objects.create(
                            question=question,
                            title=content,
                            is_correct=opt['is_correct']
                        )
                        options_created += 1
        
        print(f"Successfully imported {questions_created} questions and {options_created} options")
        
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Run the import function
if __name__ == "__main__":
    import_questions()
```

```python
import re
import json
from docx import Document

def clean_text(text):
    # Remove LaTeX-style formatting
    text = re.sub(r'\$.*?\$', lambda m: m.group().replace('$', ''), text)
    # Remove HTML comments
    text = re.sub(r'<!--.*?-->', '', text)
    # Remove other special formatting
    text = text.replace('{=html}', '')
    text = text.replace('\\', '')
    return text.strip()

def extract_questions(content):
    questions = []
    current_question = None
    current_options = []
    correct_answer = None
    
    lines = content.split('\n')
    for line in lines:
        line = clean_text(line)
        
        # Skip empty lines
        if not line.strip():
            continue
            
        # Check for question number
        question_match = re.match(r'^(\d+)\.(.+)', line)
        if question_match:
            # Save previous question if exists
            if current_question:
                questions.append({
                    "question": current_question,
                    "image": None,
                    "options": current_options,
                })
            
            # Start new question
            current_question = question_match.group(2).strip()
            current_options = []
            continue
            
        # Check for options
        option_match = re.match(r'^[#]?([A-D])[)\.]?\s*(.+)', line)
        if option_match:
            is_correct = line.startswith('#')
            option_letter = option_match.group(1)
            option_text = option_match.group(2).strip()
            
            current_options.append({
                option_letter: option_text,
                "image": None,
                "is_correct": is_correct
            })
    
    # Add the last question
    if current_question:
        questions.append({
            "question": current_question,
            "image": None,
            "options": current_options
        })
    
    return questions

def save_to_json(questions, filename):
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(questions, f, ensure_ascii=False, indent=2)

# Extract content from your document
def process_math_test(docx_path):
    # Read the Word document
    doc = Document(docx_path)
    
    # Extract text from the document
    full_text = []
    for paragraph in doc.paragraphs:
        full_text.append(paragraph.text)
    
    content = '\n'.join(full_text)
    
    questions = extract_questions(content)
    save_to_json(questions, 'math_test_5-6.json')
    return len(questions)

# Get the document content
try:
    num_questions = process_math_test('5-6 math.docx')
    print(f"Successfully processed {num_questions} questions")
except Exception as e:
    print(f"Error processing document: {str(e)}")
```

```python
import docx
import re
import json
import argparse
from typing import List, Dict, Any, Optional, Tuple

def extract_problems(document_path: str) -> List[Dict[str, Any]]:
    """
    Extract math problems from a Word document and convert to JSON with LaTeX notation.
    
    Args:
        document_path: Path to the .docx file
        
    Returns:
        List of problem dictionaries ready for JSON serialization
    """
    # Load the document
    doc = docx.Document(document_path)
    
    problems = []
    current_problem = None
    current_options = []
    correct_option = None
    
    # Regular expressions for parsing
    problem_pattern = re.compile(r'^\s*(\d+)\.\s*(.+)$')
    option_pattern = re.compile(r'^\s*([A-D])\)\s*(.+?)\s*(#)?$')
    
    for paragraph in doc.paragraphs:
        text = paragraph.text.strip()
        if not text:
            continue
        
        # Check if this is a new problem
        problem_match = problem_pattern.match(text)
        if problem_match:
            # Save previous problem if exists
            if current_problem and current_options:
                problems.append(create_problem_dict(current_problem, current_options, correct_option))
                current_options = []
                correct_option = None
            
            # Start new problem
            _, problem_text = problem_match.groups()
            current_problem = problem_text
        else:
            # Check if this is an option
            option_match = option_pattern.match(text)
            if option_match and current_problem:
                option_letter, option_text, is_correct = option_match.groups()
                current_options.append((option_letter, option_text))
                if is_correct:
                    correct_option = option_letter
    
    # Add the last problem
    if current_problem and current_options:
        problems.append(create_problem_dict(current_problem, current_options, correct_option))
    
    return problems

def create_problem_dict(problem_text: str, options: List[Tuple[str, str]], correct_option: Optional[str]) -> Dict[str, Any]:
    """
    Create a dictionary for a single problem with LaTeX notation.
    
    Args:
        problem_text: The text of the problem
        options: List of (letter, text) tuples for each option
        correct_option: The letter of the correct option
        
    Returns:
        Dictionary representation of the problem
    """
    # Convert to LaTeX notation
    problem_latex = convert_to_latex(problem_text)
    
    # Create options list
    options_list = []
    for letter, text in options:
        option_latex = convert_to_latex(text)
        options_list.append({
            letter: option_latex,
            "image": None,
            "is_correct": letter == correct_option
        })
    
    return {
        "question": problem_latex,
        "image": None,
        "options": options_list
    }

def convert_to_latex(text: str) -> str:
    """
    Convert mathematical expressions to LaTeX notation.
    
    Args:
        text: Text with potential mathematical expressions
        
    Returns:
        Text with LaTeX notation for mathematical expressions
    """
    # Replace common mathematical patterns with LaTeX
    # Fractions like a/b
    text = re.sub(r'(\d+)/(\d+)', r'\\frac{\1}{\2}', text)
    
    # Superscripts like x^2
    text = re.sub(r'(\w+)\^(\d+)', r'\1^{\2}', text)
    
    # Basic math operations
    text = re.sub(r'(\d+)\s*×\s*(\w+)', r'\1 \\times \2', text)
    text = re.sub(r'(\d+)\s*:\s*(\d+)', r'\1 : \2', text)
    
    # Check if the text contains mathematical expressions
    # If it contains operation symbols or other indicators, wrap in $
    math_indicators = ['\\frac', '\\times', ':', '=', '<', '>', '+', '-', '/', '\\cdot', '^']
    needs_math_delimiters = any(indicator in text for indicator in math_indicators)
    
    if needs_math_delimiters:
        # Make sure we're not double-wrapping already LaTeX-formatted text
        if not (text.startswith('$') and text.endswith('$')):
            text = f'${text}$'
    
    return text

def main():
    # Hardcoded input filename
    input_file = "9-10-11 test.docx"
    
    # Generate output filename by replacing the extension with .json
    output_file = input_file.rsplit('.', 1)[0] + '.json'
    
    try:
        problems = extract_problems(input_file)
        
        # Write to JSON file
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(problems, f, ensure_ascii=False, indent=2)
        
        print(f"Successfully converted {len(problems)} problems to {output_file}")
    
    except Exception as e:
        print(f"Error: {e}")

main()
```

```python
import re
import json
from pathlib import Path

def convert_math_test_to_json(content):
    """
    Convert a math test document to JSON format with proper LaTeX handling.
    """
    # Split the content by questions (starting with number followed by period)
    questions_raw = re.split(r'\n\s*\*\*(\d+)\.\*\*\s*', content)
    
    # First element is header text, remove it
    header = questions_raw[0].strip()
    # The rest alternates between question numbers and question content
    question_data = []
    
    for i in range(1, len(questions_raw), 2):
        if i+1 >= len(questions_raw):
            break
            
        question_number = questions_raw[i]
        question_content = questions_raw[i+1].strip()
        
        # Extract question text and options
        options_match = re.split(r'\n\s*([A-E]\))\s*', question_content)
        
        question_text = options_match[0].strip()
        options = {}
        
        # Process options
        for j in range(1, len(options_match), 2):
            if j+1 >= len(options_match):
                break
                
            option_letter = options_match[j].replace(')', '')
            option_text = options_match[j+1].strip()
            
            # Convert image references to LaTeX placeholders
            if "![](media/image" in option_text:
                option_text = "[Image representation not available]"
            
            # Convert fractions and mathematical expressions to LaTeX
            option_text = convert_to_latex(option_text)
            options[option_letter] = option_text
        
        # Convert question text to LaTeX
        question_text = convert_to_latex(question_text)
        
        # Remove image references from question text
        if "![](media/image" in question_text:
            question_text = re.sub(r'!\[\]\(media/image\d+\.wmf\)', "[Mathematical expression]", question_text)
        
        question_data.append({
            "number": int(question_number),
            "text": question_text,
            "options": options
        })
    
    # Create the final JSON structure
    test_json = {
        "title": "Mathematics Test",
        "questions": question_data
    }
    
    return test_json

def convert_to_latex(text):
    """
    Convert mathematical expressions to LaTeX format.
    """
    # Already in LaTeX format (within $ signs)
    if '$' in text:
        return text
    
    # Convert fraction notation
    text = re.sub(r'(\d+)/(\d+)', r'\\frac{\1}{\2}', text)
    
    # Convert exponents
    text = re.sub(r'(\w+)\^(\w+)\^', r'\1^{\2}', text)
    
    # Convert square roots
    text = re.sub(r'\\sqrt{([^}]+)}', r'\\sqrt{\1}', text)
    
    # Convert bullet points to LaTeX
    text = text.replace('•', '\\bullet')
    
    return text

def save_to_json(data, output_file):
    """
    Save data to a JSON file with proper formatting.
    """
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
    
    return f"Saved successfully to {output_file}"

# Example usage with the document content from your input


from docx import Document

def extract_text_from_docx(docx_path):
    doc = Document(docx_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

# Then use:


if __name__ == "__main__":
    # Use the document content directly instead of reading from a file
    document_content = extract_text_from_docx('9-10-11 test.docx')

    output_file = "math_test.json"
    
    # Convert to JSON
    test_json = convert_math_test_to_json(document_content)
    
    # Save to file
    result = save_to_json(test_json, output_file)
    
    print(result)
    
    # Preview the first two questions
    print("\nSample of converted content:")
    for question in test_json["questions"][:2]:
        print(f"\nQuestion {question['number']}: {question['text']}")
        for option, text in question["options"].items():
            print(f"  {option}) {text}")
```
